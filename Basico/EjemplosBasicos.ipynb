```python
import os
import ipywidgets as widgets
from IPython.display import display, clear_output
from pyswip import Prolog

# Initialize Prolog
prolog = Prolog()

# Load Prolog file
prolog.consult("duelo.pl")

# Widgets
query_input = widgets.Text(
    description='Query:',
    placeholder='e.g., parent(X, doe)'
)

output_area = widgets.Output()

# Function to query facts
def on_query_fact(change):
    query = query_input.value
    result = list(prolog.query(query))
    with output_area:
        clear_output()
        print(f"Query result: {result}")

# Button
query_button = widgets.Button(description="Query Fact")
query_button.on_click(on_query_fact)

# Display widgets
display(query_input, query_button, output_area)
```

```python
from pyswip import Prolog
import ipywidgets as widgets
from IPython.display import display, clear_output
import os

# Determine the notebook directory and path to Prolog file
notebook_dir = os.path.abspath('')
prolog_file = os.path.join(notebook_dir, 'duelo.pl')

# Initialize Prolog instance
prolog = Prolog()

# Load Prolog file
try:
    with open(prolog_file, 'r') as file:
        prolog_str = file.read()
    prolog.assertz(prolog_str)
    print(f"Prolog file '{prolog_file}' loaded successfully.")
except Exception as e:
    print(f"Error loading Prolog file '{prolog_file}': {e}")

# Widgets
query_input = widgets.Text(
    description='Query:',
    placeholder='e.g., parent(X, doe)'
)
output_area = widgets.Output()

# Function to handle querying Prolog
def on_query_fact(change):
    query = query_input.value
    try:
        result = list(prolog.query(query))
        with output_area:
            clear_output()
            print(f"Query result: {result}")
    except Exception as e:
        with output_area:
            clear_output()
            print(f"Error: {str(e)}")

# Button
query_button = widgets.Button(description="Query Fact")
query_button.on_click(on_query_fact)

# Display widgets
display(query_input, query_button, output_area)
```
